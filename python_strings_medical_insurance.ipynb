```python
medical_data = \
"""Marina Allison   ,27   ,   31.1 , 
#7010.0   ;Markus Valdez   ,   30, 
22.4,   #4050.0 ;Connie Ballard ,43 
,   25.3 , #12060.0 ;Darnell Weber   
,   35   , 20.6   , #7500.0;
Sylvie Charles   ,22, 22.1 
,#3022.0   ;   Vinay Padilla,24,   
26.9 ,#4620.0 ;Meredith Santiago, 51   , 
29.3 ,#16330.0;   Andre Mccarty, 
19,22.7 , #2900.0 ; 
Lorena Hodson ,65, 33.1 , #19370.0; 
Isaac Vu ,34, 24.8,   #7045.0"""

# Add your code here
#          ----work with string-----

#make currency readable
updated_medical_data = medical_data.replace("#", "$")
#print(updated_medical_data)

#add up the total records found
num_records = 0
for index in updated_medical_data:
  if index == "$":
    num_records += 1
#print("There are {} medical records in the data.".format(num_records))

#split string of records into list of each record
medical_data_split = updated_medical_data.split(';')
#print(medical_data_split)

#splitting each record into its own list
medical_records = []
for item in medical_data_split:
  medical_records.append(item.split(','))
#print(medical_records)

#     ------Cleaning data------

#clean up whitespace and change names to uppercase
medical_records_clean = []
for record in medical_records:
  record_clean = []
  for element in record:
    record_clean.append(element.strip())
  medical_records_clean.append(record_clean)
#print(medical_records_clean)
for record in medical_records_clean:
  record[0] = record[0].upper()
  #print(record)

#Store name, age, BMI, and insurance cost in separate lists.
names = []
ages = []
bmis = []
insurance_costs = []

for record in medical_records_clean:
  name, age, bmi, cost = record
  names.append(name)
  ages.append(age)
  bmis.append(bmi)
  insurance_costs.append(cost)

# print(names)
# print(ages)
# print(bmis)
# print(insurance_costs)

#     ----Analyze cleaned data----

#Average BMI
total_bmi = 0
for bmi in bmis:
  total_bmi += float(bmi)
average_bmi = round(total_bmi / len(bmis),2)
print("Average BMI: {bmi}".format(bmi=average_bmi))

#Average Insurance_cost
#remove '$'
costs = []
total_cost = 0
for cost in insurance_costs:
  costs.append(cost[1:])
for cost in costs:
  total_cost += float(cost)
average_cost = round(total_cost / len(insurance_costs),2)
print("Average Insurance Cost: ${cost}".format(cost=average_cost))
```
